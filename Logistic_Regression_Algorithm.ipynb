```
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset (Hours studied vs Pass/Fail)
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Initialize parameters
w = 0
b = 0
learning_rate = 0.01
epochs = 1000

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training using Gradient Descent
for i in range(epochs):
    z = w * X + b
    y_pred = sigmoid(z)

    # Gradients
    dw = np.mean((y_pred - y) * X)
    db = np.mean(y_pred - y)

    # Update parameters
    w -= learning_rate * dw
    b -= learning_rate * db

print("Weight:", w)
print("Bias:", b)

# Plot sigmoid curve
x_values = np.linspace(0, 10, 100)
z = w * x_values + b
sigmoid_curve = sigmoid(z)

plt.scatter(X, y, color='red', label='Data Points')
plt.plot(x_values, sigmoid_curve, color='blue', label='Sigmoid Curve')
plt.xlabel("Input (Hours Studied)")
plt.ylabel("Probability")
plt.title("Logistic Regression")
plt.legend()
plt.show()
```
